```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/depression_dataset_reddit_cleaned.tsv', delimiter = '\t')
```

```python
# library to clean data
import re
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
corpus = []
```

```python
# 1000 (reviews) rows to clean
for i in range(0, 7731):

    # column : "Review", row ith
    review = re.sub('[^a-zA-Z]', ' ', dataset['clean_text'][i])

    # convert all cases to lower cases
    review = review.lower()

    # split to array(default delimiter is " ")
    review = review.split()

    # creating PorterStemmer object to
    # take main stem of each word
    ps = PorterStemmer()

    # loop for stemming each word
    # in string array at ith row
    review = [ps.stem(word) for word in review
                if not word in set(stopwords.words('english'))]

    # rejoin all string array elements
    # to create back into a string
    review = ' '.join(review)

    # append each string to create
    # array of clean text
    corpus.append(review)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer

# To extract max 1500 feature.
# "max_features" is attribute to
# experiment with to get better results
cv = CountVectorizer(max_features = 1000)

# X contains corpus (dependent variable)
X = cv.fit_transform(corpus).toarray()

# y contains answers if review
# is positive or negative
y = dataset.iloc[:, 1].values
```

```python
X.shape
```

```python
y.shape
```

```python
# Splitting the dataset into
# the Training set and Test set
from sklearn.model_selection import train_test_split

# experiment with "test_size"
# to get better results
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
# Fitting Random Forest Classification
# to the Training set
from sklearn.ensemble import RandomForestClassifier

# n_estimators can be said as number of
# trees, experiment with n_estimators
# to get better results
model = RandomForestClassifier(n_estimators = 501,
                            criterion = 'entropy')

model.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)

y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

cm
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score,classification_report, accuracy_score
ran_for_pre=precision_score(y_test, y_pred)
ran_for_rec=recall_score(y_test, y_pred)
ran_for_f1=f1_score(y_test, y_pred)
ran_for_clas=classification_report(y_test, y_pred)
ran_for_acc=accuracy_score(y_test, y_pred)
print('precision: ',ran_for_pre)
print('recall: ',ran_for_rec)
print('f1: ',ran_for_f1)
print('classification: ',ran_for_clas)
print('accuracy: ',ran_for_acc)
```
